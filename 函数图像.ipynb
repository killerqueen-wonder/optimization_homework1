```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 定义函数  Define the function 
def f(x, y):
    return x**4+2*(x-y)*x*x+4*y*y
# 创建x, y的数据点 Create x, y data points 
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
x, y = np.meshgrid(x, y)

# 计算z的值 calculate the z
z = f(x, y)

# 创建一个图形 create a figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 绘制表面 create a surface
surf = ax.plot_surface(x, y, z, cmap='viridis')

plt.scatter(0, 0, color='blue', s=50)
plt.scatter(-2, 1, color='blue', s=50)

# 添加颜色条 add a color bar
fig.colorbar(surf)

# 设置轴标签 set up labels of axis
ax.set_xlabel('X1 ')
ax.set_ylabel('X2 ')
ax.set_zlabel('f3(x) ')

# 显示图形 display the figure
plt.show()
```

```python
# 创建等高线图 creat a contour 
plt.figure()
contour = plt.contour(x, y, z,levels=20)
plt.colorbar(contour)

plt.scatter(0, 0, color='red', s=100)
plt.scatter(-2, 1, color='red', s=100)
# 显示图形 display the figure
plt.show()
```
